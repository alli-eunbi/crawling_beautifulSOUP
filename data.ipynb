```python
import pandas as pd

data = pd.read_csv("final_recipe.csv")
steps = data["cooking_step"]
imgs = data["cooking_image"]
```

```python
count = 0
import ast
for i in range(len(steps)):
    stepss = ast.literal_eval(steps[i])
    imgss = ast.literal_eval(imgs[i])
    if len(imgss)>1:
        if len(stepss) != len(imgss):
            print(stepss, len(stepss))
            print(imgss, len(imgss))
            count = count+1
print(count)
```
